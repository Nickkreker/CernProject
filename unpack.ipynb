```python
import os
import tarfile
```

```python
def unpack(path_to_pack):
    folders = os.listdir(path_to_pack)
    for folder in folders:
        for i in range(1,11):
            try:
                tar = tarfile.open(f'{path_to_pack}/{folder}/jobresult_{i}.tar.gz')
                try:
                    tar.extractall(os.path.join(os.pardir, f'test_data/{folder}/jobresult_{i}'))
                except:
                    print(f'{folder}/jobresult_{i}/ SMWR')
                tar.close()
            except tarfile.ReadError:
                print(f'{folder}/jobresult_{i} Error reading file')
    
unpack('D:/CernDataset/PACK_100_199/')
```

```python
def unpack_whole_dataset(path_to_dataset, path_where_extract, start_from=0):
    packs = os.listdir(path_to_dataset)
    for pack in packs:
        path_to_pack = os.path.join(path_to_dataset, pack)
        folders = os.listdir(path_to_pack)
        for folder in folders:
            if int(folder) < start_from:
                continue
                
            for i in range(1,11):
                try:
                    tar = tarfile.open(f'{path_to_pack}/{folder}/jobresult_{i}.tar.gz')
                    try:
                        tar.extractall(os.path.join(path_where_extract, f'{folder}/jobresult_{i}'))
                    except:
                        print(f'{folder}/jobresult_{i}/ SMWR')
                    tar.close()
                except:
                    print(f'{folder}/jobresult_{i} Error reading file')

# This takes much time        
unpack_whole_dataset('C:/CernDataset/', 'D:/CernDataset/', 247)
```
